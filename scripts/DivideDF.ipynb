```r
# github.com/belisards/coronabr/

library(tidyverse)

df <- read.csv("../dados/serie_ivis.csv") %>% mutate(date = as.Date(date)) 

df
```

```r
coronabr <- df %>% select(date,time,cases,deaths,uf) %>% arrange(date)


br <- read.csv("../dados/corona_brasil.csv") %>% mutate(date = as.Date(date)) %>% filter(date > '2020-03-18')

br$time <- NA

geral <- rbind(br,coronabr) %>% arrange(date)

write_csv(geral,"../dados/corona_brasil.csv")
```
